```python
from google.colab.patches import cv2_imshow
img = cv2.imread("images.jpg")
#cv2.imshow("img",img)
cv2_imshow(img)
```

```python
face_landmark_list = face_recognition.face_landmarks(img)
face_landmark =face_landmark_list[0]
rgb_img =cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
pil_img =Image.fromarray(rgb_img)
fill_value=(158,63,136,100)
fill_value1=(128,0,128,100)
fill_value2=(0,0,0,100)
draw = ImageDraw.Draw(pil_img,'RGBA')
draw.polygon(face_landmark['left_eyebrow'],fill= fill_value1)
draw.polygon(face_landmark['right_eyebrow'],fill= fill_value1)
draw.polygon(face_landmark['left_eye'],fill= fill_value2)
draw.polygon(face_landmark['right_eye'],fill= fill_value2)
draw.polygon(face_landmark['top_lip'],fill= fill_value1)
draw.polygon(face_landmark['bottom_lip'],fill= fill_value1)
pil_img

```
